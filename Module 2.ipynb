```python
secret = 6
guess = int(input("Choose a number between 1-10"))
if guess < secret:
    print("too low")
elif guess >= secret:
    print("too high")
elif guess == secret:
    print("just right")
else:
    print("Error")
            
```

```python
small = True
green = False
if small and green:
    print("pea")
elif small and not green:
    print("cherry")
elif not small and green:
    print("watermelon")
else:
    print("pumpkin")
```

```python
guess_me = 7
number = 1
while True:
    if number < guess_me:
        print("too low")
    elif number == guess_me:
        print("found it!")
        break
    else:
        print("oops")
        break
    number += 1
print()
```
