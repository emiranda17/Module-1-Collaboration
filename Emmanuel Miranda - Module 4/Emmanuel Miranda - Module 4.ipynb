```python
import csv

rows = [
    ['author', 'book'],
    ['J R R Tolkien', 'The Hobbit'],
    ['Lynne Truss', 'Eats, Shoots & Leaves']
]

with open('books.csv', 'w', newline='') as file:
    writer = csv.writer(file, quoting=csv.QUOTE_MINIMAL)
    writer.writerows(rows)
```

```python
import sqlite3

conn = sqlite3.connect('books.db')
cur = conn.cursor()

cur.execute("""
CREATE TABLE books (
    title TEXT,
    author TEXT,
    year INTEGER
)
""")

conn.commit()
conn.close()
```
